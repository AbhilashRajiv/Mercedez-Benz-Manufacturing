```python
train.drop("ID",axis=1,inplace=True)
test.drop("ID",axis=1,inplace=True)
```

```python
l=[]
for column in train:
    for val in train[column]:
        if ((type(val)!=str) and (train[column].var()==0.0)):
            l.append(column)
            break
# checking if variance is zero in any column               
```

```python
for val in l:
    train.drop(val,axis=1,inplace=True)
    test.drop(val,axis=1,inplace=True)
    
# filtering out columns with zero variance
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def label_encoding(df):
    le=LabelEncoder()
    for column in df:
            for val in df[column]:
                if (type(val)==str):
                    df[column]=le.fit_transform(df[column])
                    break

# Apllying Label Encoder
```

```python
label_encoding(train)
```

```python
from sklearn.preprocessing import Normalizer
Nz=Normalizer()
train.loc[:,"X0":"X8"]=Nz.fit_transform(train.loc[:,"X0":"X8"])

# Normalizing Values
```

```python
train_features=train.drop("y",axis=1)
train_features
```

```python
train_target=train["y"]
train_target
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents_train = pca.fit_transform(train_features)
principalDf_train = pd.DataFrame(data = principalComponents_train
             , columns = ['principal component 1', 'principal component 2'])

# Apllying PCA for dimensionality reduction
```

```python
principalDf_train
```

```python
finalDf = pd.concat([principalDf_train, train[['y']]], axis = 1)
finalDf
```

```python
principalComponents_test = pca.transform(test)
principalDf_test = pd.DataFrame(data = principalComponents_test
             , columns = ['principal component 1', 'principal component 2'])
principalDf_test
```

```python
pca.explained_variance_ratio_
```

```python
import xgboost as xgb

# Model building using XGboost
```

```python
regressor = xgb.XGBRegressor(
    n_estimators=100,
    reg_lambda=1,
    gamma=0,
    max_depth=3,learning_rate=.1
)
```

```python
regressor.fit(principalDf_train,train_target)
```

```python
y_test_pred = regressor.predict(principalDf_test)
y_test_pred

# Final predicted Values after applying XGboost
```
